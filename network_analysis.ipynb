## 1. 수업내용 요약 <a class="anchor" id="chapter1"></a>

#### 1.1 네트워크 이론의 개요
- 네트워크: 각각의 객체들이 상호 연결된 구조
- 구성요소: node(Node, vertex)와 링크(link, edge)
- 노드: 고유한 속성을 가지고 있는 행위자
- 링크: 노드들간의 관계(relation)

#### 1.2 그래프의 기본개념
- 그래프 이론(Graph theory): 객체들 간의 네트워크 관계 모델을 수학적으로 표현된 그래프로 나타내는 연구 분야. 그래프는 노드(Node, vertex)로 구성, 이러한 노드들은 링크(Link, edge)로 연결되어 있음.
- 무방향(Undirected) 네트워크: 책, 논문의 공저자, 네트워크, 페이스북의 친구 관계 네트워크 등
- 방향(Directed) 네트워크: 인터넷 주소의 하이퍼링크 네트워크, 전화통화 네트워크, 트위터의 리트윗, 팔
로잉 네트워크 등

#### 1.3 그래프의 표현 방법
##### 가. 인접행렬(Adjacency Matrix)
* 인접한 노드들 간의 연결 관계를 행렬(Matrix)로 표현, 네트워크 상에서 노드의 개수가 n인 경우 인접 행렬은 nxn의 정방 행렬로 표현
* 무방향 네트워크의 경우 모든 관계가 쌍방향이어서 방향성이 상쇄되는 경우와 동일하게 됨. 즉 인접행렬은 대칭적인 특징을 가짐

##### 나. 노드 리스트(Node List)
* 인접한 링크가 있는 노들만을 표기하는 방법 으로 노드들만의 목록.

##### 다. 엣지 리스트(Edge List)
* 노드들을 연결한 링크들의 리스트로 “링크의 목록"
* 각각의 행은 링크를 통해서 연결되는 두 개의 노드 쌍으로 이루어져 있음. 따라서 리스트는 2열로만 이루어지게 되며, 행의 개수는 전체 링크의 개수와 일치하게 됨
* 엣지 리스트에서는 그래프의 방향성을 자체적으로 구분하기 어렵기 때문에 그래프의 방향성 여부를 별도로 명시해 주어야 함.
* 고립된 노드의 경우 링크가 존재하지 않으므로 목록에서 제외됨.

#### 1.4 네트워크 유형
- 방향성과 가중치에 따른 분류
- 관계의 위상에 따른 분류
- 네트워크 크기에 따른 분류
- 분석 대상에 따른 분류

#### 1.5 소셜 네트워크 분석 지표

##### 1.5.1 네트워크 수준의 속성 분석
- 네트워크 크기(Network Size): 네트워크를 구성하는 노드들의 수
- 연결관계의 수
    * 방향성이 있는 경우: nx(n-1)
    * 방향성이 없는 경우: nx(n-1)/2
- 네트워크 밀도(Network Density)
    * 무방향 이진 네트워크의 밀도: k/{n x (n-1)/2}
    * 방향 이진 네트워크의 밀도: k/{n x (n-1)}
    * k: 실제 연결된 링크의 수, n: 노드의 수
- 포괄성(Inclusiveness): 전체 네트워크의 밀집도를 연결된 노드의 관점에서 표현
    * n-n_isolated / n
    
##### 1.5.2 노드 수준의 속성 분석
- 연결정도(Degree): 해당 노드에 직접 연결되어 있는 노드들의 개수, 또는 링크의 개수.
- 연결거리(Distance): 노드간 관계(링크)의 강도
    - 이진 네트워크(Binary Network)
    - 가중 네트워크(Valued Network)
- 직경(Diameter)
- 평균연결거리(Average Distance)

##### 1.5.3 네트워크에 내재된 속성 분석
- 상호성(Reciprocity)
- 군집화 계수(Clustering Coefficent)

##### 1.5.4 중심성 지표
- 중심성 지표는 소셜 네트워크 분석 지표 중에서 일반적으로 가장 많이 사용되는 지표로 한 행위자가 전체 네트워크에서 중심에 위치하는 정도로 표현하는 지표
- 연결정도 중심성(Degree centrality): 한 노드가 얼마나 많은 다른 노드들과 연결관계를맺고 있는 지 측정
- 근접 중심성(Closeness centrality): 한 노드가 얼마나 전체 네트워크의 중심에 근접해 있는지를 측정
- 매개 중심성(Betweenness centrality): 한 노드가 다른 노드들 간의 네트워크 관계 형성에 있어서 중개자 혹은 매개자 역할을 얼마나 수행하는 지 측정하는 지표
- 아이겐 벡터 중심성(Eigenvector centrality): 연결된 다른 노드들의 중요도를 함께 반영한 것

```python
import numpy as np
import pandas as pd
import re
import networkx as nx
import matplotlib.pyplot as plt
```

```python
lonely_death = pd.read_excel('고독사_news.xlsx')
```

```python
lonely_death.head()
```

```python
from konlpy.tag import Hannanum
from konlpy.tag import Okt
from konlpy.tag import Kkma
```

```python
hannanum = Hannanum()
okt=Okt()
tokenizer=Kkma()
```

```python
# dataset = []
# for i in range(len(lonely_death)):
#     dataset.append(tokenizer.nouns(lonely_death['본문'][i]))
```

```python
dataset = []
for i in range(len(lonely_death)):
    dataset.append(lonely_death['키워드'][i].split(','))
```

```python
dataset[:1]
```

```python
# from mlxtend.preprocessing import TransactionEncoder
# from mlxtend.frequent_patterns import apriori, association_rules
from apyori import apriori
```

```python
# te = TransactionEncoder()
# te_ary = te.fit(dataset).transform(dataset)

assosciation_rules=apriori(dataset,min_support=0.2)
```

```python
assosciation_results=list(assosciation_rules)
```

```python
df = pd.DataFrame(assosciation_results)
```

```python
df
```

```python
df['length'] = df['items'].apply(lambda x: len(x))
df = df[(df['length'] == 2) & (df['support'] >= 0.01)].sort_values(by='support', ascending=False)
```

```python
G = nx.Graph()
```

```python
ar = (df['items']); G.add_edges_from(ar)
```

```python
pr = nx.pagerank(G)
nsize = np.array([v for v in pr.values()])
```

```python
nsize = 2000 * (nsize-min(nsize)) / (max(nsize)-min(nsize))
```

```python
pos = nx.random_layout(G)
```

```python
import matplotlib.font_manager as fm
from matplotlib import rc
font_name = fm.FontProperties(fname="c:/fonts/NanumPen.ttf").get_name()
rc('font', family='NanumPen')
```

```python
plt.figure(figsize=(16, 12));plt.axis('off')

nx.draw_networkx(G, font_family=font_name, font_size=22, pos=pos, node_color=list(pr.values()), node_size=nsize, alpha=0.7, edge_color = '.5', cmap=plt.cm.YlGn)
```
